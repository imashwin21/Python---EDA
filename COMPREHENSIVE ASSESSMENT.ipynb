# Identify the predominant age group among employees

```python
print(data['Age'].min())
print(data['Age'].max())
```

```python
# define age groups with min age 18, max age 50

bins = [18,25,30,35,40,50]
labels = ['18-25','26-30','31-35','35-40','41-50']

#define age group column

data['Age group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=True)
data['Age group']
```

```python
agegroup_count = data['Age group'].value_counts()
agegroup_count
```

```python
#predominant age group
predom = agegroup_count.idxmax()
predom
```

```python
total = agegroup_count.max()
total
```

```python
yticks = list(range(0,211,10))
```

```python
agegroup_count.plot(kind='bar',yticks = yticks, figsize =(8,5),ylabel='Number of employees'
                    ,title = 'Number of employees Vs Age group',color = 'orange')
plt.show()
```

```python
print('Age group',predom,'has',total,'employees and they are the predominant age group in the company')
```

# Discover which team and position have the highest salary expenditure 

```python
pos_salary = data[['Position','Salary']]
pos_salary
```

```python
payments = pos_salary.groupby('Position')['Salary'].sum()
payments
```

```python
positions = payments.index.tolist()
salaries = payments.values.tolist()
```

```python
plt.pie(salaries,labels=positions,colors=('r','y','g','b','m'),explode = (0.2,0,0,0,0),autopct ='%1.2f%%')
plt.show()
```

```python
highest = payments.max()
highest_exp = payments.idxmax()
print(highest_exp,'has the highest expenditure of',highest)
```

#  Investigate if there's any correlation between age and salary, and represent it visually

```python
data['Salary'].idxmax()
data.loc[[109]]
```

```python
age_count = data['Age'].value_counts()
age_count
```

```python
age_salary = data[['Age','Salary']]
age_salaries = age_salary.groupby("Age")['Salary'].mean()
age_salaries_df = pd.DataFrame(age_salaries).reset_index()
age_salaries_df['Salary'] = age_salaries_df['Salary'].astype(int) #converting float to integer for the column
age_salaries_df
```

```python
age_salaries_df.set_index('Age',inplace = True)
xticks = list(range(18,42,2))
age_salaries_df.plot(kind='line', figsize =(8,5),xticks = xticks,ylabel='Average Salary'
                    ,title = 'Salary & Age comparison',color = 'red')
plt.show()
```

#
As the age increases experience increases hence more salary is expected by the employees. Once the peak age is over salary decreases as the performing years are ending for the employees. The only exception is when the employee possesses hight market value due to the exceptional maintenance of performance. In this data, it is only one player named Koby Bryant who is paid the highest salary of all the active employees.
